```python
import pandas as pd
from sqlalchemy import create_engine
import json
```

```python
#Store CSV into DataFrame
csv_file = "Resources/us_household_income.csv"
us_household_income_df = pd.read_csv(csv_file,encoding="cp1252")
us_household_income_df.head()
```

```python
 #Create new data with select columns
new_us_household_income_df = us_household_income_df[['State_Name', 'County', 'Mean']].copy()
new_us_household_income_df.head()   
```

```python
excel_file = "Resources/earnings.xlsx"
earnings_df = pd.read_excel(excel_file, header=1)
earnings_df.head()
```

```python
earnings_edit_df = earnings_df.drop(earnings_df.index[0])
```

```python
 #Create new data with select columns
earnings_edit_df = earnings_edit_df[['State', '2014-2015']].copy()
earnings_edit_df.head() 
```

```python
earnings_new_df = earnings_edit_df.rename(columns={'2014-2015':'Median'})
```

```python
earnings_new_df = earnings_new_df[['State', 'Median']].copy()
earnings_new_df.head()
```

```python
#Store CSV into DataFrame
csv_file = "Resources/unemployment_rate.csv"
unemployment_rate_df = pd.read_csv(csv_file,encoding="cp1252")
unemployment_rate_df.head()
```

```python
#Clean DataFrame
new_unemployment_rate_df = unemployment_rate_df[["Year", "State", "County", "Rate"]].copy()
new_unemployment_rate_df.head()
```

```python
# Connect to local database
rds_connection_string = "postgres:Tenisi1@localhost:5432/ETL_Project"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
#Check for tables
engine.table_names()
```

```python
#Use pandas to load csv converted DataFrame into database
new_us_household_income_df.to_sql(name='us_household_income', con=engine, if_exists='replace', index=False)
```

```python
# Use pandas to load csv converted DataFrame into database
new_unemployment_rate_df.to_sql(name='unemployment_rate', con=engine, if_exists='replace', index=False)
```

```python
# Use pandas to load csv converted DataFrame into database
earnings_new_df.to_sql(name='earnings', con=engine, if_exists='replace', index=False)
```
